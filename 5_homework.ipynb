# Normal

Проверить гипотезу о том, что женщины выживают чаще мужчин с уровнем значимости 0.05
Проверить гипотезу о том, что дети выживают чаще, чем взрослые

```python
from scipy.stats import t
from scipy.stats import norm
```

```python
n = df.shape[0]
alpha = 0.05
```

```python
man =df[df['Sex'] == 'male']
man_serv = man['Survived']#.shape[0]/df.shape[0]
```

```python
wom =df[df['Sex'] == 'female']
wom_serv = wom['Survived']
wom_serv = wom_serv#.shape[0]/df.shape[0]
```

Гипотеза о более высокой выживаемости женщин, чем мужчин:
wom_serv > man_serv 

```python
wom_serv
```

```python
man_serv
```

```python
stat, pval = scipy.stats.ttest_ind(  wom_serv, man_serv)
critical_value = t.ppf(1 - alpha / 2, n - 1)
if abs(stat) < critical_value:
    print('H0 справедлива, более высокая выживаемости у женщин ')
else:
    print('H1 справедлива, более высокая выживаемости у мужчин')
```

```python
result = scipy.stats.ttest_1samp(wom_serv, man_serv)
result
```
